```python
#!pip install mlxtend
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
import csv
with open('groceries.csv', 'r') as f:
    data = list(csv.reader(f, delimiter=","))

data = np.array(data, dtype=object)
data
```

```python
te = TransactionEncoder()
transformado = te.fit_transform(data)
transformado
```

```python
te.columns_
```

```python
pd.set_option('display.max_columns', None)
df = pd.DataFrame(data=transformado, columns=te.columns_)
df
```

```python
qtd = 0
for i in range(0,df.shape[1]):
    if(df.iloc[i,:].value_counts()[True] == 1):
        qtd += 1
print(qtd)
```

```python
itemsets = apriori(df, min_support=0.02, use_colnames=True)
itemsets
```

```python
regras = association_rules(itemsets, metric="confidence", min_threshold=0.4)
regras
```

```python
df_regras = pd.DataFrame(regras)
```

```python
df_regras.sort_values(by='lift', ascending=False)
```
